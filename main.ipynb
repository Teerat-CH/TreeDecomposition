```python
from Graph import Graph
from Tree import Tree
from Bag import Bag
from TreeDecomposition import TreeDecomposition
from EliminationOrder import DynamicMinimumDegree
import copy
```

```python
with open('./data.txt', 'r') as file:
    data = file.readlines()
```

```python
graph = Graph()
for line in data:
  values = line.strip().split(" ")
  graph.addNode(int(values[1]))
  graph.addNode(int(values[2]))
  graph.connectNodes(int(values[1]), int(values[2]), float(values[3]))
```

```python
graph.draw()
```

```python
eliminationOrder = DynamicMinimumDegree(copy.deepcopy(graph))
```

```python
tree = TreeDecomposition(copy.deepcopy(graph), eliminationOrder)
```

```python
tree.getTreeWidth()
```

```python
tree.draw()
```
